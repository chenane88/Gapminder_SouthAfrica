```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
here='./gapminder_tidy_clean.csv'
df=pd.read_csv(here)
df.head(5)
```

```python
sa_df=df.loc[df['Country']=='South Africa']
sa_df
```

```python
sa_df.set_index('Year',inplace=True)
```

```python
sa_df.head(2)
```

```python
plt.figure(figsize=(12,10))
sns.lineplot(data=sa_df['population'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population over the years')
plt.show()
```

<b>We see that South Africa has experienced an increase in population over the years</b>

```python
plt.figure(figsize=(12,10))
sns.lineplot(data=sa_df['child_mortality'])
plt.xlabel('Year')
plt.ylabel('Chil Mortality')
plt.title('Child Mortality over the years')
```

<b>We see that there has been a reduction in child mortality over the years</b></br>
<b> The number did slightly do up between 1990 and 2000 before again reducing</b>

```python
plt.figure(figsize=(12,10))
sns.lineplot(data=sa_df['gdp'])
plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('GDP in South Africa over the years')
```

<b>GDP has been increasing towards the 2000s</b>

<b>Comparison with Kenya which attained its independece earlier</b>

```python
kenya_df=df.loc[df['Country']=='Kenya']
kenya_df.head(2)
```

```python
kenya_df.set_index('Year',inplace=True)
```

```python
plt.figure(figsize=(12,10))
sns.lineplot(data=sa_df['population'],color='blue')
sns.lineplot(data=kenya_df['population'],color='orange')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population over the years')
plt.show()
```

<b>South Africa has a larger population compared to Kenya with Kenya's growing rapidly staring the 2000s</b>

```python
plt.figure(figsize=(12,10))
sns.lineplot(data=sa_df['child_mortality'],color='blue')
sns.lineplot(data=kenya_df['child_mortality'],color='orange')
plt.xlabel('Year')
plt.ylabel('Child Mortality')
plt.title('Child Mortality over the years')
plt.show()
```

<b>In the beginning Kenya had a lower child mortality rate compared to South Africa but this has changed with the passage of time</b>

```python
plt.figure(figsize=(12,10))
sns.lineplot(data=sa_df['gdp'],color='blue')
sns.lineplot(data=kenya_df['gdp'],color='orange')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('GDP over the years')
plt.show()
```

<b>South Africa GDP is higher than Kenya's by too large a margin</b>

```python
plt.figure(figsize=(12,10))
sns.lineplot(data=sa_df['fertility'],color='blue')
sns.lineplot(data=kenya_df['fertility'],color='orange')
plt.xlabel('Year')
plt.ylabel('Fertility')
plt.title('Fertility over the years')
plt.show()
```

<b>Kenya has a higher ferility rate than South Africa</b></br>
<b> During the comparison between their child mortality rates we saw Kenya having a higher child mortality rate and hence this could explain the difference in population</b>

<b>The region in question is the sub-saharan region</b>

```python
Africa_df=df.loc[df['region']=='Sub-Saharan Africa']
Africa_df
```

```python
year_df=Africa_df.query('Year==2013')
```

<b>Ten Largest GDPs of 2013 in the Sub-Saharan Region</b>

```python
plt.figure(figsize=(12,10))
sns.barplot(x='Country',y='gdp',data=year_df.nlargest(10,columns='gdp'),palette='inferno')
plt.title('Ten Largest GDPs of 2013')
plt.xticks(rotation=90)
```

<b>South Africa has the 6th Largest GDP in the SubSaharan region</b>

```python
year_df.set_index('Country',inplace=True)
```
